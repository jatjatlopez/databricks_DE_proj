#Transformation Logic

```python
query = """
SELECT
    ROW_NUMBER() OVER (ORDER BY pn.product_start_date, pn.product_key) AS product_key, -- Surrogate key
    pn.product_id,
    pn.product_key AS product_number,
    pn.product_name,
    pn.cat_id,
    pc.category,
    pc.subcategory,
    pc.maintainance_flag AS maintenance_flag,
    pn.product_line,
    pn.product_start_date AS start_date
FROM silver.crm_products pn
LEFT JOIN silver.erp_product_category pc
    ON pn.cat_id = pc.category_id
--WHERE pn.end_date IS NULL; -- Filter out all historical data
"""
df = spark.sql(query)

df.limit(10).display()

     
```

#Writing to Gold Layer

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_products")
```
