#Ingest Files to Bronze Layer

```python

for path in base_path:
    print(f'Ingesting {path['source']} to workspace.bronze.{path['table']}')

    df = spark.read.option('header','true').option('inferSchema','true').csv(path['path'])
    
    df.write.mode('overwrite').format('delta').saveAsTable(f'workspace.bronze.{path['table']}')
```
