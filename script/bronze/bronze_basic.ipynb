# Reading from CSV File

```python
df1 = (
    spark.read.option('header', 'True')
    .option('inferSchema', 'True')
    .csv('/Volumes/workspace/bronze/source_systems/cust_info.csv')
)

df2 = (
    spark.read.option('header', 'True')
    .option('inferSchema', 'True')
    .csv('/Volumes/workspace/bronze/source_systems/prd_info.csv')
)

df3 = (
    spark.read.option('header', 'True')
    .option('inferSchema', 'True')
    .csv('/Volumes/workspace/bronze/source_systems/sales_details.csv')
)

df4 = (
    spark.read.option('header', 'True')
    .option('inferSchema', 'True')
    .csv('/Volumes/workspace/bronze/source_systems/CUST_AZ12.csv')
)

df5 = (
    spark.read.option('header', 'True')
    .option('inferSchema', 'True')
    .csv('/Volumes/workspace/bronze/source_systems/LOC_A101.csv')
)

df6 = (
    spark.read.option('header', 'True')
    .option('inferSchema', 'True')
    .csv('/Volumes/workspace/bronze/source_systems/PX_CAT_G1V2.csv')
)

display(df3)
```

#Write it to Bronze Layer

```python
df1.write.mode('overwrite').option('overwriteSchema', 'true').saveAsTable('workspace.bronze.crm_cust_info')
df2.write.mode('overwrite').option('overwriteSchema', 'true').saveAsTable('workspace.bronze.crm_prd_info')
df3.write.mode('overwrite').option('overwriteSchema', 'true').saveAsTable('workspace.bronze.crm_sales_details')
df4.write.mode('overwrite').option('overwriteSchema', 'true').saveAsTable('workspace.bronze.erp_cust_az12')
df5.write.mode('overwrite').option('overwriteSchema', 'true').saveAsTable('workspace.bronze.erp_loc_a101')
df6.write.mode('overwrite').option('overwriteSchema', 'true').saveAsTable('workspace.bronze.erp_px_cat_g1v2')
```
